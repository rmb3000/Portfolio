```python
import requests
import pandas as pd

# CMS Open Payments API - 2023 General Payments
base_url = "https://openpaymentsdata.cms.gov/api/1/datastore/query/fb3a65aa-c901-4a38-a813-b04b00dfa2a9/0"

# Tell the API not to compress the response
headers = {
    "Accept-Encoding": "identity"
}

params = {
    "limit": 500,
    "offset": 0,
    "count": "true"
}

print("Connecting to CMS Open Payments API...")
response = requests.get(base_url, params=params, headers=headers)
print(f"Status: {response.status_code}")

if response.status_code == 200:
    data = response.json()
    print(f"Total records available: {data.get('count', 'unknown')}")
    df_bronze = pd.DataFrame(data['results'])
    print(f"Records retrieved: {len(df_bronze)}")
    print(df_bronze.columns.tolist())
    display(df_bronze.head(3))
else:
    print(f"Error: {response.text}")
```

```python
# Write Bronze data to Lakehouse Files as Parquet
df_bronze.to_parquet("/lakehouse/default/Files/bronze_cms_payments.parquet", index=False)
print(f"Successfully written {len(df_bronze)} records to Bronze Lakehouse")
print("Location: Files/bronze_cms_payments.parquet")
```
